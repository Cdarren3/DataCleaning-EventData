```python
df['Date (DD/MM/YY)']=df[0].apply(lambda x : re.findall(r'(?<=Date\t).*2018', x)[0])
df['No. of Event Days']=df[0].apply(lambda x : 2 if re.findall(r'(?<=2018\s)(\-).*2018', x) else 1)
df['Name of Event (English)']=df[0].apply(lambda x : re.findall(r'.*(?=\nDate)', x)[0])
df['Name of Event (Chinese)']=df[0].apply(lambda x : x.replace(x, ""))
#for data that is not exist in every event detail
df['Promoter']=df[0].apply(lambda x : re.findall(r'(?<=Organiser\t).*', x))
df['Promoter']=df['Promoter'].apply(lambda x : '' if len(x)==0 else x[0])

df['Event Type 1']=df[0].apply(lambda x : x.replace(x, ""))
df['Event Type 2']=df[0].apply(lambda x : x.replace(x, ""))
df['Importance']=df[0].apply(lambda x : x.replace(x, ""))
df['Nature']=df[0].apply(lambda x : x.replace(x, ""))
#quick sort by the data in "Date" column
df['Frequency 1']=df[0].apply(lambda x : 'Recurrent' if re.findall(r'(?<=2018\s)(\-).*2018', x) else '')
df['Frequency 2']=df[0].apply(lambda x : 'Times/Year' if re.findall(r'(?<=2018\s)(\-).*2018', x) else '')
#for data that is not exist in every event detail
df['Admission Charges']=df[0].apply(lambda x : re.findall(r'(?<=Entrance Fee\t).*', x))
df['Admission Charges']=df['Admission Charges'].apply(lambda x : '' if len(x)==0 else x[0])

df['No. of Spectators']=df[0].apply(lambda x : x.replace(x, ""))
df['Gross ticket proceeds']=df[0].apply(lambda x : x.replace(x, ""))
#for data that is not exist in every event detail
df['Remarks']=df[0].apply(lambda x : re.findall(r'(?<=Remarks\t)([\s\S]*)(?=\n.+\t)', x))
df['Remarks']=df['Remarks'].apply(lambda x : '' if len(x)==0 else x[0])

df['Venue Type']=df[0].apply(lambda x : re.findall(r'(?<=Venue\t).*', x)[0])

```
